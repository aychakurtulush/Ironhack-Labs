```python
def clean_gender(x):
    if x in ['M', 'Male']:
        return 'Male'
    elif x in ['F', 'female',"Femal"]:
        return 'Female'
    elif np.nan:  pass
    else:
        return 'U'
```

```python
df['gender'] = list(map(clean_gender, df['gender'])) 
```
