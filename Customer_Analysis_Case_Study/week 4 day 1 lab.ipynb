Devide the numerical and categorical columns seperately

```python
numerical_columns = df.select_dtypes(include=['int', 'float']).columns
numerical = df[numerical_columns]
numerical
```

```python
categorical_columns = df.select_dtypes(include=['object']).columns
categoricals = df[categorical_columns]
categoricals
```

Check distributions of numerical variables

```python
for col in numerical.columns:
    sns.displot(numerical[col])
    plt.show()
```

Check correlation between numerical variables and see the correlation heatmap

```python
corr = numerical.corr()
```

```python
sns.heatmap(corr, 
        xticklabels=corr.columns,
        yticklabels=corr.columns, 
        annot=True, cmap='Reds')
```

PowerTransformer usage for our second variable that has high level of skewness

```python
pt = PowerTransformer(method='yeo-johnson') #yeo-johnson method usage
```

```python
numerical['total_claim_amount']= pt.fit_transform(numerical[['total_claim_amount']])
```

```python
round(numerical['total_claim_amount'].mean()) #check for the mean
```
