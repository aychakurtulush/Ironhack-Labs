Converting Grams/Mile to Grams/Km

Converting Gallons to Liters

What brand has the most cars?

```python
### your code us here
most_common_brand = cars['Model'].value_counts().idxmax()
print(most_common_brand)
```

<b>show the average CO2_Emission_Grams/Km  by Brand

```python
### your code us here

cars['CO2 Emission Grams/Mile'] = cars['CO2 Emission Grams/Mile'] * 1.60934
cars.rename(columns={'CO2 Emission Grams/Mile': 'CO2 Emission Grams/Km'}, inplace=True)

cars.groupby(['Make']).mean()
```

# (Optional) 

Use `pd.cut` or `pd.qcut` to create 4 groups (bins) of cars, by Year. We want to explore how cars have evolved decade by decade.

```python
cars['Year'].value_counts()
```

```python
# your code here
labels = ['1984 - 1992', '1993 - 2001', '2002 - 2010', '2011 - 2016']

# Use the cut function to create a new column with the categories
cars['cat_year'] = pd.qcut(cars['Year'], q=4, labels=labels)

cars['cat_year'].value_counts()
```
